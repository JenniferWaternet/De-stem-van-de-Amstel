# Openai key + Open Webui

Voor het gebruik van taalmodellen hebben we voor elk team een openai account en key gemaakt. Deze kun je vinden op het openai platform (https://platform.openai.com/docs/overview). Deze key kan je in je code zetten of om het makkelijker te maken hebben we een guide gemaakt voor het opzetten van een interface waardoor je met een llm kan communiceren. Dit is een opensourced interface van [Open WebUI](https://github.com/open-webui/open-webui). Deze notebook gaat over het opzetten van de open webui omgeving en hoe je je openai key kan gebruiken en eventueel ook andere opensourced modellen kan inladen.

## Opzetten webui omgeving
1. Installeer [Docker desktop](https://www.docker.com/products/docker-desktop/)
2. Start docker desktop

*Optioneel toevoegen Ollama modellen*

3. *Download [ollama](https://ollama.com/)*
4. *Open cmd en voeg een ollama model toe*

```
ollama pull <model-name>
```
5. Open cmd en run command

```
docker run -d -p 3000:8080 --add-host=host.docker.internal:host-gateway -v open-webui:/app/backend/data --name open-webui --restart always ghcr.io/open-webui/open-webui:main
```

4. Open in een browser [localhost:3000](http://localhost:3000/)
5. Ga naar de beheerderspaneel



6. Ga naar instellingen -> verbindingen -> Beheer OpenAI API-verbindingen -> en vul daar je groeps openai key in.



7. Selecteer een model (of meerdere) en experimenteer erop los!
